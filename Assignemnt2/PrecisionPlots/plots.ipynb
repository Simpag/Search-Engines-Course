```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rankedfile = r"C:\Users\Simon\Documents\Github\Search-Engines-Course\Assignemnt2\search-engine\assignment-2-4.txt"
rankedScores = []
with open(rankedfile) as f:
    lines = f.readlines()
    assert len(lines) == 50, "Wrong number of lines! " + str(len(lines))
    rankedScores = [float(line.strip().split(" ")[-1]) for line in lines]
```

```python
def calc_precision(relevant, total):
    return relevant/total

def calc_recall(relevant, numDocs):
    return relevant/numDocs
```

```python
numDocs = 100
rankedPrecisions = []
rankedRecalls = []
for i in [10,20,30,40,50]:
    s = np.array(rankedScores[:i])
    s[s>0] = 1
    rankedPrecisions.append(calc_precision(np.sum(s), len(s)))
    rankedRecalls.append(calc_recall(np.sum(s), numDocs))
```

```python
plt.plot(rankedRecalls, rankedPrecisions)
for i in range(len(rankedRecalls)):
    plt.plot(rankedRecalls[i], rankedPrecisions[i], 'o', label=f"N={10*(i+1)}")
plt.grid()
plt.title("Ranked")
plt.legend()
plt.xlabel("Recall")
plt.ylabel("Precision")
```
